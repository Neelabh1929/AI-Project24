Linear Regression

```python
#Write Code for Logistic Regression here
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


input_data = cancer_dataset[["Age","BMI","Glucose","Insulin","HOMA","Leptin","Adiponectin","Resistin","MCP.1"]]
target = cancer_dataset["Classification"]
x_train, x_test, y_train, y_test = train_test_split(input_data, target,test_size=0.15)
model = LogisticRegression(max_iter=100000)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_of_logistic = accuracy_score(y_test, y_pred)
print("Accuracy of Logistic Regression:", accuracy_of_logistic*100)
```

SVM(Support Vector Machines)

```python
#Write Code for SVM here
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


input_data = cancer_dataset[["Age","BMI","Glucose","Insulin","HOMA","Leptin","Adiponectin","Resistin","MCP.1"]]
target = cancer_dataset["Classification"]
x_train, x_test, y_train, y_test = train_test_split(input_data, target, test_size=0.3)
model = SVC()
model.fit(x_train, y_train)
y_pred_svm = model.predict(x_test)
acc = accuracy_score(y_test, y_pred_svm)
print("Accuracy of SVM is:", acc*100)
```

# **LINEAR REGRESSION**

```python
#For more information for the dataser refer: https://archive.ics.uci.edu/dataset/477/real+estate+valuation+data+set
real_estate_dataset = pd.read_excel('/content/drive/MyDrive/Real estate valuation data set.xlsx')
```

```python
real_estate_dataset.head()
```

```python
# @title X1 transaction date vs X2 house age

from matplotlib import pyplot as plt
real_estate_dataset.plot(kind='scatter', x='X1 transaction date', y='X2 house age', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
real_estate_dataset.tail()
```

```python
#Code for Linear Regression here

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error


columns = ["X1 transaction date", "X2 house age", "X3 distance to the nearest MRT station", "X4 number of convenience stores", "X5 latitude", "X6 longitude"]
req = "Y house price of unit area"
X = real_estate_dataset[columns]
y = real_estate_dataset[req]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
#print(y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute error is: ",mae)
```
